```python
import numpy as np
import matplotlib.pyplot as plt

# Group the data by 'Transactor_revolver' and calculate the average activity for each time period
grouped_data = df.groupby('Transactor_revolver')[['T+1_month_activity', 'T+2_month_activity', 'T+3_month_activity', 'T+6_month_activity', 'T+12_month_activity']].mean()

# Calculate the percentage for each value in the grouped data
total_sum = grouped_data.sum(axis=1)
percentage_data = grouped_data.divide(total_sum, axis=0) * 100

# Transpose the percentage data for plotting
percentage_data = percentage_data.T

# Define the x-axis labels
x_labels = ['T+1 Month', 'T+2 Months', 'T+3 Months', 'T+6 Months', 'T+12 Months']

# Get the number of bars and width for each bar
n_bars = percentage_data.shape[1]
bar_width = 0.35

# Set the index for positioning the bars
index = np.arange(len(x_labels))

# Plot the bars
fig, ax = plt.subplots(figsize=(10, 6))
for i, col in enumerate(percentage_data.columns):
    plt.bar(index + i * bar_width, percentage_data[col], bar_width, label=col)

    # Add percentage labels to each bar
    for j, value in enumerate(percentage_data[col]):
        plt.text(index[j] + i * bar_width, value + 1, f'{value:.1f}%', ha='center')

# Set the x-axis ticks and labels
plt.xticks(index + (n_bars - 1) * bar_width / 2, x_labels)

# Set the y-axis label
plt.ylabel('Percentage')

# Set the plot title
plt.title('Percentage of Average Activity by Time Period and Transactor Revolver')

# Add a legend
plt.legend()

# Display the plot
plt.show()

```
